# <span style="color:#FFAAAA"> Hill Climbing</span>

```python
def hillclimb(start, get_neighbors, heur):
    current = start

    while True:
        n = get_neighbors(current)

        if not n:
            return current
        
        best = max(n, key=lambda x: heur(x))

        if heur(best) <= heur(current):
            return current
        
        current = best

graph = {
    'A' : [('B', 1), ('C', 3)],
    'B' : [('D', 1)],
    'C' : [('D', 1)],
    'D' : []
}

def get_neightbors(node):
    return graph[node]

def heuristic(node):
    h = {'A': 3, 'B' : 2, 'C': 1, 'D': 0}
    return h[node]


```

# <span style="color:#FFAAAA">Beam Search</span>

```python
def beamsearch(start, goal, get_neighbors, heur, beamwidth):
    beam = [start]

    while beam:
        if goal in beam:
            return "Goal found"
        
        newbeam = []

        for node in newbeam:
            newbeam.extend(get_neighbors(node))

        newbeam.sort(key=lambda x: heur(x), reverse=True)
        beam = newbeam[:beamwidht]
    return "No solution"
```

# <span style="color:#FFAAAA">Genetic Algo</span>

```python
import random

def genetic(population, fitness, crossover, mutate, generations):

    for i in range(generations):
        population = sorted(population, key=lambda x: fitness(x["genes"]), reverse=True)
        parent1, parent2 = population[0], population[1]
        newpopulation = []

        for _ in range(len(population)):
            child = crossover(parent1,parent2)
            child = mutate(child)
            newpopulation.append(child)
        
        population = newpopulation
    return max(population, key=lambda x: fitness(x["genes"]))

#fitness function for n-queens
#index = value
#row = col
solution = [1,3,0,2]

def create_population(size, n):
    population = []

    for _ in range(size):
        individual = list(range(n))
        random.shuffle(individual)
        population.append({
            "genes":individual,
            "parents": None
            })
    return population

def fitness(solution):
    n = len(solution)
    nonattacking = 0
    for i in range(n):
        for j in range(i+1, n):
            if solution[i] != solution[j] and abs(solution[i] - solution[j]) != abs(i-j):
                nonattacking +=1
    return nonattacking

def mutate(individual):
    genes = individual["genes"]
    n = len(genes)

    i,j = random.sample(range(n), 2)
    genes[i], genes[j] = genes[j], genes[i]
    return individual

def crossover(p1, p2):
    parent1 = p1["genes"]
    parent2 = p2["genes"]

    n = len(parent1)
    point = random.randint(1, n - 2)

    child = parent1[:point]

    for gene in parent2:
        if gene not in child:
            child.append(gene)

    return {
        "genes": child,
        "parents": (parent1, parent2)
    }

def printlineage(individual):
    print("Current: ", individual["genes"])
    if individual["parents"]:
        print("Parents:")
        print("P1: ", individual["parents"][0])
        print("P2: ", individual["parents"][1])
    else:
        print("No parents(initial population)")
#main for n-queens
n=4
popsize = 5
generations = 50
population = create_population(popsize,n)

result = genetic(population, fitness, crossover, mutate, generations)

print("Best Solution: ", result["genes"])
print("Fitness: ", fitness(result["genes"]))

printlineage(result)
```
