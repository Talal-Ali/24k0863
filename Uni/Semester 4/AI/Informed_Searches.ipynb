# <span style="color:#FFAAAA">Best-First Search</span>

```python
import heapq


def bfs(start, goal, get_neighbors, heur):
    open_list = []
    heapq.heappush(open_list, (heur(start,goal), start))

    visited = set()

    while open_list:
        priority, current = heapq.heappop(open_list)
        if current == goal:
            return "Goal found"
        visited.add(current)

        for n in get_neighbors(current):
            if n not in visited:
                heapq.heappush(open_list, (heur(n, goal), n))
    
    return "No Solution"


#Helpers

graph = {
    'A' : [('B', 1), ('C', 3)],
    'B' : [('D', 1)],
    'C' : [('D', 1)],
    'D' : []
}

def get_neightbors(node):
    return graph[node]

def heuristic(node, goal):
    h = {'A': 3, 'B' : 2, 'C': 1, 'D': 0}
    return h[node]

#main
```

# <span style="color:#FFAAAA">Greedy BFS</span>

```python
import heapq

def gbfs(start, goal, get_neightbors, heur):
    pq = []
    heapq.heappush(pq, (heur(start, goal), start))
    parent = {start: None}
    visited = set()

    while pq:
        priority, node = heapq.heappop(pq)

        if node == goal:
            path = []
            while node:
                path.append(node)
                node = parent[node]
            return path[::-1]
        visited.add(node)

        for n, _ in get_neightbors(node):
            if n not in visited:
                parent[n] = node
                heapq.heappush(pq, (heur(n,goal), n))
    return "No Solution"
#Helpers

graph = {
    'A' : [('B', 1), ('C', 3)],
    'B' : [('D', 1)],
    'C' : [('D', 1)],
    'D' : []
}

def get_neightbors(node):
    return graph[node]

def heuristic(node, goal):
    h = {'A': 3, 'B' : 2, 'C': 1, 'D': 0}
    return h[node]

print(gbfs('A', 'D', get_neightbors, heuristic))
```

# <span style ="color:#FFAAAA">A* Search</span>

```python
import heapq

def astar(start, goal, get_neightbors, heur):
    open_list = []
    heapq.heappush(open_list, (0,start))

    gcost = {start:0}

    parent = {start:None}
    visited = set()

    while open_list:
        priority, current = heapq.heappop(open_list)
        if current == goal:
            path = []
            while current:
                path.append(current)
                current = parent[current]
            return path[::-1]
        
        visited.add(current)

        for n, cost in get_neightbors(current):
            newcost = gcost[current] + cost

            if n not in gcost or newcost<gcost[n]:
                gcost[n] = newcost
                parent[n] = current
                f = newcost + heur(n,goal)
                heapq.heappush(open_list, (f,n))
    return "No Solution"
#Helpers

graph = {
    'A' : [('B', 1), ('C', 3)],
    'B' : [('D', 1)],
    'C' : [('D', 1)],
    'D' : []
}

def get_neightbors(node):
    return graph[node]

def heuristic(node, goal):
    h = {'A': 3, 'B' : 2, 'C': 1, 'D': 0}
    return h[node]

print(astar('A', 'D', get_neightbors, heuristic))
```
