# <span style="color:#FFBB77">Simple Reflexive Agent</span>

```python
import random
class Light:
    def __init__(self, inten = "dim", mot = "no", switch = "off"):
        self.inten = inten
        self.mot = mot
        self.switch = switch

    def status(self):
        print("Intensity : " + self.inten + " | Motion: " + self.mot + " | Switch: " + self.switch)

    def turnon(self, motion, intensity):
        if intensity == "bright":
            self.switch = "off"
        if motion == "yes" and intensity == "dim":
            self.switch = "on"
        elif motion == "no" and intensity == "dim":
            self.switch = "off"
        self.mot = motion
        self.inten = intensity
        self.status()
        
agent = Light()
for i in range(10):
    intensity = random.choice(["bright", "dim"])
    motion = random.choice(["yes", "no"])
    agent.turnon(motion, intensity)
```

# <span style="color:#FFBB77">Normal BFS & DFS</span>

```python
tree = {
    'A' : ['B', 'C'],
    'B' : ['D', 'E'],
    'C' : ['E'],
    'D' : []

}

def bfs(graph, start, end):
    visited = []
    queue = []
    visited.append(start)
    queue.append(start)

    while queue:
        node = queue.pop(0)
        print(node, " ")
        if(node == end):
            print("Goal found")
            break
        for n in graph[node]:
            if n not in visited:
                visited.append(n)
                queue.append(n)
def dfs(graph, start, end):
    visited = []
    stack = []
    visited.append(start)
    stack.append(start)

    while stack:
        node = stack.pop(-1)
        print(node, " ")
        if node == end:
            print("Goal found")
            break
        for n in reversed(graph[node]):
            if n not in visited:
                visited.append(n)
                stack.append(n)


start = 'A'
end = 'D'

bfs(tree, start, end)
dfs(tree, start, end)
```

# <span style ="color:#34BAFF">UCS</span>

```python
import heapq

graph = {
    'A' : [('B', 9), ('C', 5)],
    'B' : [('D', 2), ('E', 1)],
    'C' : [('E', 2), ('B' , 3)],
    'D' : []

}

def ucs(graph, start, goal):
    queue = [(0, start, [])]
    visited = set()

    while queue:
        cost, node, path = heapq.heappop(queue)

        if node not in visited:
            
            visited.add(node)
            path = path + [node]

            if node == goal:
                return path,cost
            
            for neighbor, weight in graph.get(node, []):
                if neighbor not in visited:
                    tcost = cost + weight
                    heapq.heappush(queue, (tcost, neighbor, path))
    return None, float('inf')

start = 'A'
goal = 'D'
path, cost = ucs(graph, start, end)
print(f"Path: {path}")
print(f"Cost: {cost}")
```

# <span style ="color:#34BAFF">DLS & IDLS</span>

```python
def dls(graph, start, goal, limit):
    stack = [(start, [start], 0)]
    while stack:
        node, path, depth = stack.pop()
        if node == goal:
            return path
        if depth< limit:
            for n in reversed(graph.get(node, [])):
                if n not in path:
                    stack.append((n, path + [n], depth+1))
    return None

def iterative_dls(graph, start, goal, maxdepth):
    for d in range(maxdepth +1):
        result = dls(graph, start, goal, d)
        if result:
            return result
    return None
```
